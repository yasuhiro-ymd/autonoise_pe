```python
from pathlib import Path
import numpy as np

# 対象フォルダを指定
folder_path = Path("../../data/artificial_noisy_v6/")

# フォルダ内の .npy ファイルを取得
npy_files = sorted([
    f for f in folder_path.glob("*.npy")
    if not f.name.startswith("._")
])

# ファイル名と shape を表示
for npy_file in npy_files:
    arr = np.load(npy_file, allow_pickle=True)
    print(f"{npy_file.name}: {arr.shape}")
```
